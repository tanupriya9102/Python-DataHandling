```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line chart using pyplot
a=[1,2,3,4]
b=[2,4,6,8]
plt.xlabel("values")
plt.ylabel("their doubles")
plt.plot(a,b)
```

```python
plt.plot(a,b,'r',marker='+',markersize=5,markeredgecolor='b')
```

```python
# scatter charts using plot()
plt.plot(a,b,'o',markersize=8)
```

```python
p=np.arange(1,20,1.25)
q=np.log(p)
r=np.cos(p)
plt.plot(p,q)
plt.plot(p,r,'r',linestyle='dashdot')
# plt.xlabel("values")
# plt.ylabel("log values")
```

```python
# scatter charts using scatter()
plt.scatter(a,b,s=12,c='m',marker='D')
```

```python
# barchats
l,m,n=[1,2,3,4],[2,4,6,8],[1,4,9,16]
plt.bar(l,m)
```

```python
plt.bar(l,n,color='orange')
```
