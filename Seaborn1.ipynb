```python
# Univariate distribution (visulaise data of single column)
sns.distplot(pollution_data['NO2.AQI'])
plt.show()
# histogram and kernel desity estimation curve(kde=smoother representation of data)
```

```python
dist=sns.distplot(pollution_data['NO2.AQI'],color="red")
dist.set_title("Pollution Stats")
dist.set_xlabel("Air quality index of NO2")
dist.set_ylabel("Density")
plt.show()
```

```python
dist=sns.distplot(pollution_data['NO2.AQI'],color="green",vertical=True)
dist.set_title("Pollution Stats")
dist.set_ylabel("Air quality index of NO2")
dist.set_xlabel("Density")
plt.show()
```

```python
dist1=sns.distplot(pollution_data["NO2.AQI"],
                kde_kws={"color":"b",
                         "linewidth":2,
                         "label":"KDE"},
                 
                 hist_kws={"histtype":"step",
                          "linewidth":2,
                          "alpha":0.7,
                          "color":"g",
                          "label":"HIST"})
dist1.set_title("Pollution stats")
dist1.set_xlabel("Air quality index of NO2")
dist1.set_ylabel("Density")
plt.show()
```

```python
dist=sns.distplot(pollution_data['NO2.AQI'],
                  color="green",
                  bins=10,
                 kde=False) #hist=false if want to see kde
dist.set_title("Pollution Stats")
dist.set_xlabel("Air quality index of NO2")
dist.set_ylabel("Frequency")
plt.show()
```

```python
rug=sns.rugplot(pollution_data['NO2.AQI'],
                 height=0.7,
                 axis="x")
rug.set_title("RUG PLOT")
rug.set_xlabel("Air quality index of NO2")
plt.show()
```

```python
plt.subplots(figsize = (20,5))
dist=sns.distplot(pollution_data['NO2.AQI'],bins=95,rug=True)
dist.set_title("HISTOGRAM,KDE & RUG plot")
dist.set_ylabel("Air quality index of NO2")
dist.set_xlabel("Density")
plt.show()
```

```python
sns.set(color_codes = True)

plot = sns.kdeplot(pollution_data["NO2.AQI"],
                   shade = True,
                   color = "b")

plot.set_title("KDE PLOT")
plot.set_xlabel("Air Quality index of NO2")
plot.set_ylabel("Density")

plt.show()
```

```python
plt.subplots(figsize = (17,7))

# bw= bandwidth

sns.kdeplot(pollution_data["NO2.AQI"] ,
            color = "k")

sns.kdeplot(pollution_data["NO2.AQI"] , 
            bw = 0.3 ,
            label = "bw:0.5",
            color = "b")

sns.kdeplot(pollution_data["NO2.AQI"] ,
            bw = 8 ,
            label = "bw:8", 
            color = "g")

sns.kdeplot(pollution_data["NO2.AQI"] ,
            bw = 30 ,
            label = "bw:30" ,
            color = "y")

sns.kdeplot(pollution_data["NO2.AQI"] ,
            bw = 80 , 
            label = "bw:80" ,
            color = "r")

plt.show()
```

```python
joint_plot = sns.jointplot("NO2.Mean",
                           "CO.Mean",
                           data = pollution_data)

plt.show()
```

```python
# regression line
joint_plot = sns.jointplot("NO2.Mean",
                           "CO.Mean",
                           data = pollution_data, 
                           kind = "reg",
                           color = "g")

plt.show()
```

```python
# set x & y limits
joint_plot = sns.jointplot("NO2.Mean",
                           "CO.Mean",
                           data = pollution_data, 
                           xlim = {0,60},
                           ylim = {0,3})

plt.show()
```

```python
# hex distribution
with sns.axes_style("darkgrid"):
    
    sns.jointplot("NO2.Mean",
                  "CO.Mean",
                   data = pollution_data,
                   kind = "hex",
                   color = "green")
    
plt.show()
```

```python
with sns.axes_style("ticks"):
    sns.jointplot("NO2.Mean",
                  "CO.Mean",
                   data = pollution_data,
                   kind = "kde",
                   space = 0,
                   color = "g")
    
plt.show()
```
