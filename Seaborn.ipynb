```python
# Univariate distribution (visulaise data of single column)
sns.distplot(pollution_data['NO2.AQI'])
plt.show()
# histogram and kernel desity estimation curve(kde=smoother representation of data)
```

```python
dist=sns.distplot(pollution_data['NO2.AQI'],color="red")
dist.set_title("Pollution Stats")
dist.set_xlabel("Air quality index of NO2")
dist.set_ylabel("Density")
plt.show()
```

```python
dist=sns.distplot(pollution_data['NO2.AQI'],color="green",vertical=True)
dist.set_title("Pollution Stats")
dist.set_ylabel("Air quality index of NO2")
dist.set_xlabel("Density")
plt.show()
```

```python
dist1=sns.distplot(pollution_data["NO2.AQI"],
                kde_kws={"color":"b",
                         "linewidth":2,
                         "label":"KDE"},
                 
                 hist_kws={"histtype":"step",
                          "linewidth":2,
                          "alpha":0.7,
                          "color":"g",
                          "label":"HIST"})
dist1.set_title("Pollution stats")
dist1.set_xlabel("Air quality index of NO2")
dist1.set_ylabel("Density")
plt.show()
```
